```python
print("=" * 50)
print("      PHISHING AWARENESS TRAINING")
print("=" * 50)

print("\nWhat is Phishing?")
print("Phishing is a cyber attack where attackers")
print("try to steal personal information using")
print("fake emails, messages, or websites.")

print("\nTips to Stay Safe:")
tips = [
    "1. Check the sender's email address.",
    "2. Don't click suspicious links.",
    "3. Verify website URLs before logging in.",
    "4. Never share OTPs or passwords.",
    "5. Enable Two-Factor Authentication (2FA)."
]

for tip in tips:
    print(tip)

print("\nQuiz Time!")
score = 0

q1 = input("1. Should you click links from unknown emails? (yes/no): ")
if q1.lower() == "no":
    print("Correct!")
    score += 1
else:
    print("Wrong! Correct answer: no")

q2 = input("2. Is HTTPS safer than HTTP? (yes/no): ")
if q2.lower() == "yes":
    print("Correct!")
    score += 1
else:
    print("Wrong! Correct answer: yes")

q3 = input("3. Should you share your OTP with anyone? (yes/no): ")
if q3.lower() == "no":
    print("Correct!")
    score += 1
else:
    print("Wrong! Correct answer: no")

print("\nYour Score:", score, "/3")

if score == 3:
    print("Excellent! You are aware of phishing attacks.")
elif score == 2:
    print("Good! Learn a little more to stay safer.")
else:
    print("Please review phishing safety tips again.")
```
